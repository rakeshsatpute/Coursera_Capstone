<h1>Part 1 - Importing the Canada postal code data from Wiki </h1>

```python
#Importing pandas and numpy libraries
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
#Importing wikipedia library
!pip install wikipedia
import wikipedia as wp
print('Wiki installed')
```

```python
#Importing postal codes data of Canada from Wiki
html = wp.page("List of postal codes of Canada: M").html().encode("UTF-8")
try:
    df = pd.read_html(html)[0]
except IndexError:
    df = pd.read_html(html)[1]

df.shape
```

```python
# Deleting rows where Boroughs are 'not assigned'
df = df[df.Borough != "Not assigned"]

# Assigning Brough to Neighbourhood where it is 'not assigned'
df['Neighborhood'] = np.where(df['Neighborhood'] == "Not assigned", df['Borough'], df['Neighborhood'])
```

<h1> Part 2 - Importing geospatial data and merging it with Canada postal code data </h1>

```python
#Importing geospatial file of Canada postal codes
file_name = 'http://cocl.us/Geospatial_data'
df_geo_data = pd.read_csv(file_name)
df_geo_data.shape
```

```python
#Joining Postal code and geospatial data
df_canada = pd.merge(df, df_geo_data, on="Postal Code")
df_canada.shape
```

#### Function to get nearby venues

```python
def getVenues(name, latitude, longitude, radius=500, LIMIT=100):
    venues_list = []
    for name, lat, lng in zip(name, latitude, longitude):
        #print(name)
        
        #API URL request
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            Client_ID, Client_Secret, Version, lat, lng, radius, LIMIT)
      
        #get request
        results = requests.get(url).json()["response"]["groups"][0]["items"]
    
        #return relevant information of each nearby venue
        venues_list.append([(
            name,
            lat,
            lng,
            v['venue']['name'],
            v['venue']['location']['lat'],
            v['venue']['location']['lng'],
            v['venue']['categories'][0]['name']) for v in results])
    
        nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
        
        nearby_venues.columns = ["Neighborhood","Neighborhood latitude","Neighborhood longitude",
                             "Venue","Venue Latitude","Venue Longitude","Venue Category"]
    return(nearby_venues)
```

#### Running the above function to get venues of Toronto neighborhoods

```python
toronto_venues = getVenues(name = df_toronto["Neighborhood"], 
                           latitude = df_toronto["Latitude"],
                           longitude = df_toronto["Longitude"])
```

#### Analyse each neighborhood

```python
#one hot encoding
toronto_onehot = pd.get_dummies(toronto_venues[["Venue Category"]], prefix="", prefix_sep="")

#add neighborhood back to dataframe
toronto_neighborhood = toronto_venues[['Neighborhood']]
toronto_onehot = toronto_neighborhood.merge(toronto_onehot, left_index=True, right_index=True)

#rnaming column neighborhood_x to neighborhood
toronto_onehot.rename(columns={"Neighborhood_x": "Neighborhood"}, inplace=True)

toronto_onehot.shape
```

#### group rows by neighborhood and take the mean of the frequency of occurrence of each category

```python
toronto_grouped = toronto_onehot.groupby('Neighborhood').mean().reset_index()
#toronto_grouped.rename(columns={"Neighborhood_x": "Neighborhood"}, inplace=True)
toronto_grouped
```

```python
toronto_grouped.shape
```

#### Function to sort venues in descending order

```python
def ret_most_cmn_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

#### New dataframe to display top 10 enues of each neighborhood

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

#create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))
        
#create a new dataframe
neighborhood_venues_sorted = pd.DataFrame(columns=columns)
neighborhood_venues_sorted['Neighborhood'] = toronto_grouped['Neighborhood']

for ind in np.arange(toronto_grouped.shape[0]):
    neighborhood_venues_sorted.iloc[ind, 1:] = ret_most_cmn_venues(toronto_grouped.iloc[ind, :], num_top_venues)
    
neighborhood_venues_sorted.head()
```

#### Clustering neighborhoods

```python
k=4

toronto_grouped_clustering = toronto_grouped.drop('Neighborhood',1)

#run k-means clustering
kmeans = KMeans(n_clusters=k, random_state=0).fit(toronto_grouped_clustering)

#check cluster labels generated for each row of the dataframe
kmeans.labels_[0:38]
```

#### Creating dataframe that includes cluster as well as top 10 venues for each neighborhood

```python
# add clustering labels
neighborhood_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

toronto_merged = df_toronto

#merge df_toronto with neighborhood_venues_sorted
toronto_merged = toronto_merged.join(neighborhood_venues_sorted.set_index('Neighborhood'), on='Neighborhood')

toronto_merged.shape
```

```python
toronto_merged.head()
```

#### Visualize in a map

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['Neighborhood'], toronto_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
